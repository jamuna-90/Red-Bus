```python
#1.APSRTC all bus route link 

def get_bus_route_details(url, num_pages=5):
    

    driver = webdriver.Chrome()
    wait = WebDriverWait(driver, 30)
    driver.get(url)

    all_data = []

    def scrape_page():
        routes_container = wait.until(EC.presence_of_all_elements_located((By.CLASS_NAME, "route_link")))

        for route in routes_container:
            try:
                route_name = route.find_element(By.TAG_NAME, 'a').text
                route_link = route.find_element(By.TAG_NAME, "a").get_attribute('href')
                all_data.append({"route_name": route_name, 'route_link': route_link})
            except Exception as e:
                print(f"An error occurred: {e}")
                continue
        book_now_buttons = driver.find_elements(By.CSS_SELECTOR, "a[class ='time-btn-30']")

        for button in book_now_buttons:
            button.click()


    # Scrape data from multiple pages
    for page_number in range(1, num_pages + 1):
        scrape_page()
        if page_number < num_pages:
            try:
                pagination_container = wait.until(EC.presence_of_element_located(
                    (By.XPATH, '//*[@id="root"]/div/div[4]/div[12]')
                ))
                next_page_button = pagination_container.find_element(
                    By.XPATH, f'.//div[contains(@class, "DC_117_pageTabs") and text()="{page_number + 1}"]'
                )

                actions = ActionChains(driver)
                actions.move_to_element(next_page_button).perform()
                time.sleep(1)
                next_page_button.click()

                wait.until(EC.text_to_be_present_in_element(
                    (By.XPATH, '//div[contains(@class, "DC_117_pageTabs DC_117_pageActive")]'), str(page_number + 1)
                ))
                time.sleep(3)
            except Exception as e:
                print(f"An error occurred while navigating to page {page_number + 1}: {e}")
                break

    driver.quit()
    return all_data

APSRTC_Routes_url ="https://www.redbus.in/online-booking/apsrtc/?utm_source=rtchometile"
apsrtc_data = get_bus_route_details(APSRTC_Routes_url)
print("APSRTC Routes:")

for route in apsrtc_data:
    route_url = route['route_link'] 
    print(route_url)


# Convert the list to a DataFrame
df = pd.DataFrame(apsrtc_data, columns=['route_name', 'route_link'])

# Save the DataFrame to a CSV file
df.to_csv("apsrtc_data.csv", index=False)
```

```python
#bus_route_details

def get_bus_details(url):
    

    driver = webdriver.Chrome()
    wait = WebDriverWait(driver, 30)
    driver.get(url)

    def scroll_to_bottom():
        body = driver.find_element(By.TAG_NAME, 'body')
        scrolling = True

        while scrolling:
            old_page_source = driver.page_source
            body.send_keys(Keys.PAGE_DOWN)
            time.sleep(2)
            new_page_source = driver.page_source

            if new_page_source == old_page_source:
                scrolling = False

    scroll_to_bottom()

    view_bus_buttons = driver.find_elements(By.CSS_SELECTOR, "a.button.view-seats")
    for button in view_bus_buttons:
        try:
            button.click()
            time.sleep(2)
        except Exception as e:
            print(f"Error clicking button: {e}")

    bus_elements = driver.find_elements(By.CLASS_NAME, "bus-item-details")
    all_bus_data = []

    for bus in bus_elements:
        try:
            bus_name = bus.find_element(By.CLASS_NAME, "travels").text
            bus_type = bus.find_element(By.CLASS_NAME, "bus-type").text
            boarding_time = bus.find_element(By.CLASS_NAME, "dp-time").text
            duration = bus.find_element(By.CLASS_NAME, "dur").text
            dropping_time = bus.find_element(By.CLASS_NAME, "bp-time").text

            try:
                rating = bus.find_element(By.CLASS_NAME, "rating-sec").text
            except:
                rating = "No rating"

            bus_info = {
                'buslink': url,
                'busname': bus_name,
                'bustype': bus_type,
                'boardingtime': boarding_time,
                'busduration': duration,
                'droppingtime': dropping_time,
                'busrating': rating
            }

            all_bus_data.append(bus_info)

        except Exception as e:
            print(f"An error occurred while extracting bus details: {e}")

    driver.quit()
    return all_bus_data

# URL for Vijayawada to Hyderabad route
vijayawada_to_hyderabad_url = "https://www.redbus.in/bus-tickets/vijayawada-to-hyderabad?fromCityId=134&toCityId=124&fromCityName=Vijayawada&toCityName=Hyderabad&busType="

# Get bus details
bus_details_data = get_bus_details(vijayawada_to_hyderabad_url)


all_apsrtc_details = []
for route in apsrtc_data:
    route_url = route['route_link']
    bus_details_for_route = get_bus_details(route_url)
    all_apsrtc_details.extend(bus_details_for_route)





for bus in all_apsrtc_details:
  print(bus)




# Convert the list to a DataFrame
df_all_apsrtc_details = pd.DataFrame(all_apsrtc_details, columns=['buslink','busname','bustype','boardingtime','busduration','droppingtime','busrating'])

# Save the DataFrame to a CSV file

df_all_apsrtc_details .to_csv("all_apsrtc_details.csv", index=False)

```

```python
#10.WBTC route details
def get_bus_route_details(url, num_pages=5):
    

    driver = webdriver.Chrome()
    wait = WebDriverWait(driver, 30)
    driver.get(url)

    all_data = []

    def scrape_page():
        routes_container = wait.until(EC.presence_of_all_elements_located((By.CLASS_NAME, "route_link")))

        for route in routes_container:
            try:
                route_name = route.find_element(By.TAG_NAME, 'a').text
                route_link = route.find_element(By.TAG_NAME, "a").get_attribute('href')
                all_data.append({"route_name": route_name, 'route_link': route_link})
            except Exception as e:
                print(f"An error occurred: {e}")
                continue
        book_now_buttons = driver.find_elements(By.CSS_SELECTOR, "a[class ='time-btn-30']")

        for button in book_now_buttons:
            button.click()


    # Scrape data from multiple pages
    for page_number in range(1, num_pages + 1):
        scrape_page()
        if page_number < num_pages:
            try:
                pagination_container = wait.until(EC.presence_of_element_located(
                    (By.XPATH, '//*[@id="root"]/div/div[4]/div[12]')
                ))
                next_page_button = pagination_container.find_element(
                    By.XPATH, f'.//div[contains(@class, "DC_117_pageTabs") and text()="{page_number + 1}"]'
                )

                actions = ActionChains(driver)
                actions.move_to_element(next_page_button).perform()
                time.sleep(1)
                next_page_button.click()

                wait.until(EC.text_to_be_present_in_element(
                    (By.XPATH, '//div[contains(@class, "DC_117_pageTabs DC_117_pageActive")]'), str(page_number + 1)
                ))
                time.sleep(3)
            except Exception as e:
                print(f"An error occurred while navigating to page {page_number + 1}: {e}")
                break

    driver.quit()
    return all_data

# Example usage:
WBTC_Routes_url ="https://www.redbus.in/online-booking/wbtc-ctc/?utm_source=rtchometile"
wbtc_data = get_bus_route_details(WBTC_Routes_url )
print("(WBTC Routes:")

for route in wbtc_data:
    route_url = route['route_link'] 
    print(route_url)


# Convert the list to a DataFrame
df_wbtc_data = pd.DataFrame(wbtc_data , columns=['route_name', 'route_link'])

# Save the DataFrame to a CSV file
df_wbtc_data.to_csv("wbtc_data .csv", index=False)
```
