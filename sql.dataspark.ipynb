```python
#customer details
mycursor.execute('''CREATE TABLE IF NOT EXISTS customer_details(
                                    CustomerKey INT,
                                    Gender VARCHAR(255),
                                    Name VARCHAR(255),
                                    City VARCHAR(100),
                                    State_code VARCHAR(255),
                                    State VARCHAR(100),
                                    Zip_Code INT,
                                    Country VARCHAR(100),
                                    Continent VARCHAR(100),
                                    Birthday DATETIME(6) )''')
mydb.commit()



for index, row in df_customer.iterrows():
     
    row = row.replace({np.nan: None})
    
    # Handle the timestamp conversion
    birthday = row["Birthday"]
    if birthday is not None:
        # Assuming birthday is a pandas Timestamp object
        birthday_str = birthday.strftime('%Y-%m-%d %H:%M:%S')  # Format for MySQL DATETIME
    else:
        birthday_str = None
    
    insert_query = '''INSERT INTO customer_details (
                          CustomerKey,
                          Gender,
                          Name,
                          City,
                          State_code,
                          State,
                          Zip_Code,
                          Country,
                          Continent,
                          Birthday)
                        VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s)'''
    

    values = (
        row["CustomerKey"],
        row["Gender"],
        row["Name"],
        row["City"],
        row["State Code"],
        row["State"],
        row["Zip Code"],
        row["Country"],
        row["Continent"],
        birthday_str)  # Use the converted string
    
    print(f"Values {index}: {values}")

    try:
        mycursor.execute(insert_query, values)
        mydb.commit()
    except mysql.connector.Error as err:
        print(f"Error: {err}")

print(mycursor.rowcount, "record(s) inserted.")
```

```python
#sales details

mycursor.execute('''CREATE TABLE IF NOT EXISTS sales_details1(
                                    Order_Number INT,
                                    Line_Item INT,
                                    Order_Date DATETIME(6),
                                    Delivery_Date DATETIME,
                                    CustomerKey INT,
                                    StoreKey INT,
                                    ProductKey INT,
                                    Quantity INT,
                                    Currency_Code VARCHAR(100) )''')
mydb.commit()


for index, row in df_sales.iterrows():


    #  datetime convert for 'Order Date' and 'Delivery Date'
    order_date = row["Order Date"]

    if order_date is not None and not pd.isna(order_date):
        order_date_str = order_date.strftime('%Y-%m-%d %H:%M:%S')
    else:
        order_date_str = None

    delivery_date = row["Delivery Date"]
    
    if delivery_date is not None and not pd.isna(delivery_date):
        delivery_date_str = delivery_date.strftime('%Y-%m-%d') 
    else:
        delivery_date_str = None

    insert_query = '''INSERT INTO sales_details1 (
                        Order_Number,
                        Line_Item,
                        Order_Date,
                        Delivery_Date,
                        CustomerKey,
                        StoreKey,
                        ProductKey,
                        Quantity,
                        Currency_Code)
                        VALUES (%s, %s, %s, %s, %s, %s, %s, %s,%s)'''
    

    values = (
        row["Order Number"],
        row["Line Item"],
        order_date_str,  
        delivery_date_str,
        row["CustomerKey"],  
        row["StoreKey"],
        row["ProductKey"],
        row["Quantity"],
        row["Currency Code"])  # Use the converted string
    
    print(f"Values {index}: {values}")

    try:
        mycursor.execute(insert_query, values)
        mydb.commit()
    except mysql.connector.Error as err:
        print(f"Error: {err}")

print(mycursor.rowcount, "record(s) inserted.")
```
